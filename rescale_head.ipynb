```python
import nibabel as nb
from nilearn import plotting
```

```python
t1w_path = "/home/pcorvila/Documents/MRI/sources/MNI02mo.nii"
```

```python
t1w_img = nb.load(t1w_path)
```

```python
plotting.plot_anat(t1w_img)
```

```python
t1w_img.header['pixdim']
```

```python
rescaled_data = t1w_img.get_fdata().copy()
```

```python
rescaled_affine = t1w_img.affine.copy()
rescaled_affine
```

```python
rescaled_affine[0:3,0:3] = 0.5*rescaled_affine[0:3,0:3]
```

```python
rescaled_affine
```

```python
rescaled_img = nb.Nifti1Image(rescaled_data, rescaled_affine, t1w_img.header)
```

```python
plotting.plot_anat(rescaled_img)
```
